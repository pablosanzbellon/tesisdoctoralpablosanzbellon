## <font color='gray'>**Preparación de los datos**</font>

El <font color='red'>**conjunto de datos**</font> (*dataset*) que vamos a utilizar en esta práctica consta de <font color='blue'>**99 radiografías de Tórax y 99 radiografías de Pelvis**</font> <font color='red'>**en formato DICOM extraídas del PACS**</font> (*Picture Archiving and Communication System*) del Hospital Universitario Marqués de Valdecilla. Las imágenes fueron completamente <font color='red'>**anonimizadas**</font> para su uso en esta práctica. 

Las imágenes fueron procesadas aplicándoles las siguientes operaciones:

*   <font color='blue'>**Inversión de los valores de píxel**</font> (p' = 4095 - p)
*   <font color='blue'>**Reescalado de los valores de píxel**</font> (p' = 1.0*p + 0.0)
*   <font color='blue'>**Aplicación de una ventana**</font> (Centro=2047, Anchura=4095)
*   <font color='blue'>**Redimensionado de la matriz de píxeles**</font> (224x224)
*   <font color='blue'>**Conversión de la imagen en escala de grises a color**</font> (RGB)
*   <font color='blue'>**Normalización de la imagen en color**</font> (p' = p/4095)

Se construyó un <font color='red'>**tensor de rango 4**</font> con todas las <font color='blue'>**imágenes procesadas**</font> de dimensión <font color='red'>**(198,224,224,3)**</font> y un <font color='red'>**tensor de rango 2**</font> con las <font color='blue'>**etiquetas**</font> de las imágenes de dimensión <font color='red'>**(198,2)**</font>.

Las imágenes de la clase <font color='blue'>**pelvis**</font> se marcaron con la etiqueta <font color='red'>**[1,0]**</font> mientras que las imágenes de la clase <font color='blue'>**tórax**</font> se marcaron con la etiqueta <font color='red'>**[0,1]**</font> (<font color='blue'>*one hot encoded*</font>).

El <font color='blue'>**conjunto de datos de entrenamiento**</font> está formado por el <font color='red'>**80%**</font> de las imágenes y el <font color='blue'>**conjunto de datos de validación**</font> está formado por el otro <font color='red'>**20%**</font> de las imágenes.

```python
#Importación de las librerías os, numpy, matplotlib, opencv y pydicom
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pydicom
#Importación del módulo que implementa el reescalado de los valores de píxel
from pydicom.pixel_data_handlers.util import apply_modality_lut
#Importación del módulo que implementa la aplicación de una ventana a los valores de píxel
from pydicom.pixel_data_handlers.util import apply_voi_lut
```

```python
#Lectura de los datos de una imagen de cada clase 
ds_pelvis = pydicom.dcmread(path+'TRAINING IMAGES/antero-posterior PELVIS/' + \
                            '1.2.826.0.1.3680043.2.135.737626.80140287.7.1596957125.14.67_1002_001002_1596957125003a.dcm')
ds_chest = pydicom.dcmread(path+'TRAINING IMAGES/postero-anterior CHEST/' + \
                           '1.2.826.0.1.3680043.2.135.737626.80140287.7.1595279102.267.91_1001_001001_15952791020026.dcm')
```

```python
#Imprimo los datos de la radiografía de pelvis
print(ds_pelvis)
```

```python
#Imprimo los datos de la radiografía de tórax
print(ds_chest)
```

```python
#Extracción de la matriz de píxeles de la radiografía de pelvis
pelvis_pixel_array = ds_pelvis.pixel_array
#Visualización de la matriz de píxeles de la radiografía de pelvis
plt.imshow(pelvis_pixel_array, cmap='gray') 
```

```python
#Extracción de la matriz de píxeles de la radiografía de tórax
chest_pixel_array = ds_chest.pixel_array
#Visualización de la matriz de píxeles de la radiografía de tórax
plt.imshow(chest_pixel_array, cmap='gray') 
```

```python
#Inversión de la matriz de píxeles de la radiografía de pelvis
print(np.power(2,ds_pelvis.BitsStored))
pelvis_pixel_array_inverted = np.power(2,ds_pelvis.BitsStored) - ds_pelvis.pixel_array
#Visualización de la matriz de píxeles de la radiografía de pelvis invertida
plt.imshow(pelvis_pixel_array_inverted, cmap='gray') 
```

```python
#Inversión de la matriz de píxeles de la radiografía de tórax
print(np.power(2,ds_pelvis.BitsStored))
chest_pixel_array_inverted = np.power(2,ds_chest.BitsStored) - ds_chest.pixel_array
#Visualización de la matriz de píxeles de la radiografía de tórax invertida
plt.imshow(chest_pixel_array_inverted, cmap='gray')
```

```python
#Reescalado de los valores de píxel de la radiografía de pelvis
pelvis_pixel_array_rescaled= apply_modality_lut(pelvis_pixel_array_inverted, ds_pelvis)
#Visualización de la matriz de píxeles de la radiografía de pelvis reescalada
plt.imshow(pelvis_pixel_array_rescaled, cmap='gray') 
```

```python
#Reescalado de los valores de píxel de la radiografía de tórax
chest_pixel_array_rescaled= apply_modality_lut(chest_pixel_array_inverted, ds_chest)
#Visualización de la matriz de píxeles de la radiografía de tórax reescalada
plt.imshow(chest_pixel_array_rescaled, cmap='gray') 
```

```python
#Aplicación de la ventana a los valores de píxel de la radiografía de pelvis
pelvis_pixel_array_windowed = apply_voi_lut(pelvis_pixel_array_rescaled, ds_pelvis)
#Visualización de la matriz de píxeles de la radiografía de pelvis después de aplicar la ventana
plt.imshow(pelvis_pixel_array_windowed, cmap='gray') 
```

```python
#Aplicación de la ventana a los valores de píxel de la radiografía de tórax
chest_pixel_array_windowed = apply_voi_lut(chest_pixel_array_rescaled, ds_chest)
#Visualización de la matriz de píxeles de la radiografía de tórax después de aplicar la ventana
plt.imshow(chest_pixel_array_windowed, cmap='gray') 
```

```python
#Impresión de las dimensiones de la matriz de píxeles de la radiografía de pelvis antes del redimensionado 
print(pelvis_pixel_array_windowed.shape)
#Redimensionado de la matriz de píxeles de la radiografía de pelvis a 224 píxeles x 224 píxeles
pelvis_pixel_array_resized = cv2.resize(pelvis_pixel_array_windowed, dsize=(224, 224), interpolation=cv2.INTER_CUBIC)
#Visualización de la matriz de píxeles de la radiografía de pelvis después de redimensionarla
plt.imshow(pelvis_pixel_array_resized, cmap='gray')
#Impresión de las dimensiones de la matriz de píxeles de la radiografía de pelvis despues del redimensionado 
print(pelvis_pixel_array_resized.shape)
```

```python
#Impresión de las dimensiones de la matriz de píxeles de la radiografía de tórax antes del redimensionado 
print(chest_pixel_array_windowed.shape)
#Redimensionado de la matriz de píxeles de la radiografía de tórax a 224 píxeles x 224 píxeles
chest_pixel_array_resized = cv2.resize(chest_pixel_array_windowed, dsize=(224, 224), interpolation=cv2.INTER_CUBIC)
#Visualización de la matriz de píxeles de la radiografía de tórax después de redimensionarla
plt.imshow(chest_pixel_array_resized, cmap='gray')
#Impresión de las dimensiones de la matriz de píxeles de la radiografía de tórax despues del redimensionado 
print(chest_pixel_array_resized.shape)
```

```python
#Conversión de la imagen en escala de grises de la radiografía de pelvis a una imagen en color
pelvis_pixel_array_rgb = np.stack((pelvis_pixel_array_resized,pelvis_pixel_array_resized,pelvis_pixel_array_resized), axis=-1)
#Impresión de las dimensiones de la imagen en color de la radiografía de pelvis
print(pelvis_pixel_array_rgb.shape)
#Normalización de la imagen en color de la radiografía de pelvis
pelvis_pixel_array_rgb_normalized = np.divide(pelvis_pixel_array_rgb,np.power(2,ds_pelvis.BitsStored))
#Visualización de la imagen en color de la radiografía de pelvis normalizada
plt.imshow(pelvis_pixel_array_rgb_normalized, cmap='gray')
```

```python
#Conversión de la imagen en escala de grises de la radiografía de tórax a una imagen en color
chest_pixel_array_rgb = np.stack((chest_pixel_array_resized,chest_pixel_array_resized,chest_pixel_array_resized), axis=-1)
#Impresión de las dimensiones de la imagen en color de la radiografía de tórax
print(chest_pixel_array_rgb.shape)
#Normalización de la imagen en color de la radiografía de tórax
chest_pixel_array_rgb_normalized = np.divide(chest_pixel_array_rgb,np.power(2,ds_chest.BitsStored))
#Visualización de la imagen en color de la radiografía de tórax normalizada
plt.imshow(chest_pixel_array_rgb_normalized, cmap='gray')
```

```python
#Definición de una función para realizar todo el procesado de las imágenes de forma automática y guardarlas en archivos binarios numpy (npy) en el directorio destino
origin = path+'TRAINING IMAGES/antero-posterior PELVIS/'
destiny = path+'TRAINING ARRAYS/antero-posterior PELVIS/'
filename = '1.2.826.0.1.3680043.2.135.737626.80140287.7.1596957125.14.67_1002_001002_1596957125003a.dcm'

def image_processing(origin,destiny,filename):
  ds = pydicom.dcmread(origin + filename)
  #print(ds)
  pixel_array = ds.pixel_array
  pixel_array_inverted = np.power(2,ds.BitsStored) - ds.pixel_array
  pixel_array_rescaled= apply_modality_lut(pixel_array_inverted, ds)
  pixel_array_windowed = apply_voi_lut(pixel_array_rescaled, ds)
  pixel_array_resized = cv2.resize(pixel_array_windowed, dsize=(224, 224), interpolation=cv2.INTER_CUBIC)
  pixel_array_rgb = np.stack((pixel_array_resized,pixel_array_resized,pixel_array_resized), axis=-1)
  pixel_array_rgb_normalized = np.divide(pixel_array_rgb,np.power(2,ds.BitsStored))
  np.save(destiny + filename[:-4] + '.npy',pixel_array_rgb_normalized)

image_processing(origin,destiny,filename)
```

```python
# Aplicación de la función que realiza el procesado a todas las imágenes de pelvis
origin = path+'TRAINING IMAGES/antero-posterior PELVIS/'
destiny = path+'TRAINING ARRAYS/antero-posterior PELVIS/'

for root, dirs, files in os.walk(origin, topdown=False):
   for name in files:
      print(os.path.join(root, name))
      image_processing(origin,destiny,name)
```

## <font color='gray'>**Construcción del modelo**</font>



### **Red neuronal convolucional** (*base_model*)

En la construcción del modelo utilizaremos la técnica de <font color='blue'>**transferencia del aprendizaje**</font> (*transfer learning*). Partimos de una red preentrenada sobre millones de imágenes del conjunto de datos de ImageNet, le quitamos el clasificador con el que se entrenó, le añadimos encima nuestro  clasificador y lo entrenamos para que reconozca imágenes médicas. La <font color='red'>**red neuronal convolucional preentrenada**</font> de la que partimos es la <font color='blue'>**VGG16**</font> <font color='red'>**con los pesos de ImageNet**</font>. VGG16 es un modelo de red neuronal convolucional propuesto por K. Simonyan y A. Zisserman de la Universidad de Oxford. El modelo alcanzó una exactitud del 92.7% en ImageNet, que es un conjunto de datos formado por 14 millones de imágenes pertenecientes a 1000 clases diferentes. La red preentrenada (<font color='blue'>*base_model*</font>) se utiliza <font color='red'>**para realizar la extracción de las características**</font> (*features*).

![VGG16](https://neurohive.io/wp-content/uploads/2018/11/vgg16-1-e1542731207177.png)




### **Clasificador** (*top_model*)

El <font color='red'>**clasificador**</font> esta formado por las siguientes capas (*layers*) y se utiliza <font color='red'>**para escoger la frontera que separa los dos clases**</font> bajo consideración (<font color='blue'>**pelvis y tórax**</font>). :

##### **Capa de agrupación** (*global average pooling layer*)

*   Realiza un <font color='red'>**promedio del valor de todos los píxeles**</font> <font color='blue'>**para cada mapa de características**</font> .

##### **Capa densa** (*fully connected layer*)

*   Con <font color='blue'>**256**</font> <font color='red'>**neuronas**</font> y una <font color='red'>**función de activación**</font> <font color='blue'>**ReLU**</font>.

![Dense](https://ichi.pro/assets/images/max/724/1*MwnCiIMMBzf2F6wr8fuSTA.jpeg)
![ReLU](https://krisbolton.com/images/posts/2018/ReLU-activation-function.jpg)

##### **Capa de descarte** (*dropout layer*)

*   Esta capa <font color='blue'>**desactivará aleatoriamente el 50% de las neuronas**</font> <font color='red'>**para que no se memoricen las imágenes de entrenamiento**</font>, no se produzca sobreajuste (*overfitting*).

![Dropout](https://miro.medium.com/max/567/1*sIWt678-YMnANnU3Gtmlag.png)

##### **Capa densa de salida** (*fully conected output layer*)
*   Con <font color='blue'>**2**</font> <font color='red'>**neuronas**</font> y una <font color='red'>**función de activación**</font> <font color='blue'>**softmax**</font>. Esta función de activación se emplea <font color='red'>**cuando tenemos múltiples clases**</font> y queremos conocer la probabilidad de pertenecer a las distintas clases posibles.

![Softmax](https://i.ytimg.com/vi/lvNdl7yg4Pg/maxresdefault.jpg)



```python
#Importación la red preentrenada (VGG16 con los pesos de Imagenet) y eliminación de las capas que conformaban el clasificador con el que se entrenó
from tensorflow.keras.applications.vgg16 import VGG16
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

```python
#Construcción del modelo a partir de la red preentrenada (base_model) que extrae las características y 
#un clasificador que asigna a cada imagen una probabilidad de pertenecer a cada clase (pelvis y tórax)
from keras.models import Sequential
from keras.layers import GlobalAveragePooling2D, Dense, Dropout
from keras.models import Model

model_top = Sequential()
model_top.add(GlobalAveragePooling2D(input_shape=base_model.output_shape[1:],
data_format=None)),
model_top.add(Dense(256, activation='relu'))
model_top.add(Dropout(0.5))
model_top.add(Dense(2, activation='softmax'))
model = Model(inputs=base_model.input, outputs=model_top(base_model.output))
```

## <font color='gray'>**Entrenamiento del modelo**</font>

##### **Función de coste** (*loss function*)
La <font color='red'>**diferencia o error**</font> <font color='blue'>**entre las predicciones de la red**</font>  (*predictions*) <font color='blue'>**y las etiquetas**</font>  (*labels*).

La **función de coste** utilizada fue la <font color='blue'>**categorical cross entropy**</font>.

![Categorical Cross Entropy](https://i.ytimg.com/vi/tRsSi_sqXjI/maxresdefault.jpg)

##### **Algoritmo de retropropagación**

El **entrenamiento de un modelo** consiste en <font color='red'>**encontrar los pesos de las neuronas que minimizan la función de coste**</font>. 

El entrenamiento de un modelo se lleva a cabo mediante el **algoritmo de retropropagación** (*backpropagation*) que <font color='red'>**consiste en cambiar los pesos en la dirección en la que la función de coste decrece más rápidamente**</font>, es decir, <font color='blue'>**en la dirección de su gradiente**</font> (**algoritmo del descenso del gradiente**). Es como ir descendiendo por un paisaje montañoso hacia un valle lo más rápidamente posible.

![Backpropagation](https://www.researchgate.net/profile/Sung-Eun-Kim-4/publication/275721804/figure/fig2/AS:651149220773890@1532257480242/Schematic-diagram-of-backpropagation-training-algorithm-and-typical-neuron-model.png)

##### **Optimizador** (*Optimizer*) 

El **optimizador** es el encargado de <font color='red'>**mejorar los pesos para que el error en la predicción sea cada vez menor**</font>. El optimizador calcula las derivadas parciales de la función de coste respecto de los pesos y modifica estos en la dirección negativa del gradiente. El optimizador utilizado fue <font color='blue'>**RMSprop**</font>.

```python
# Compilación del modelo con un optimizador Adam y una función de coste categorical crossentropy
# La tasa de aprendizaje (learning rate) escogida fue pequeña (0.0001) para facilitar la transferencia del aprendizaje
# El resto de los parámetros del optimizador toman los valores por defecto
# La métrica empleada fue la exactitud (accuracy) 

from keras.optimizers import RMSprop
model.compile(optimizer=RMSprop(learning_rate=0.00001), loss='categorical_crossentropy', metrics=['accuracy'])
```
